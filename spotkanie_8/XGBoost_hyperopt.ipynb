```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from scipy import stats
from scipy.stats import skew
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

#wyświetl wszystkie wiersze i kolumny
pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('dataset_pm25.csv',index_col = 0, header = 0)
df.head()
```

```python
#Usuwany niepotrzebne stacje
df.drop(['Kondratowicza', 'MzWarMarszal-PM2.5-1g', 'Wokalna', 'MzWarTolstoj-PM2.5-1g'], axis = 1, inplace = True)
df.head()
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Suma', 'Udział'])
missing_data
```

```python
#Usuwam cechy, które mają powyżej 11% brakujących dancyh
df = df.drop((missing_data[missing_data['Udział'] > 0.110556]).index,1)
df.head()
```

```python
#Usuwam kolumny zawierające kody, statusy pomiaru
cols = df[['Kod_stacji', 'Nazwa_stacji', 'Wysokość_podstawy_tekstowy_[opis]', 'Wysokość_podstawy_chmur_CL_CM_szyfrowana_[kod]', 'Widzialność_[kod]', 'Status_pomiaru_PORW', 'Wskaźnik_wentylacji_[W/N]', 'Status_pomiaru_TTZW', 'Charakterystyka_tendencji_[kod]', 'Rodzaj_opadu_za_6_godzin_[kod]', 'Pogoda_bieżąca_[kod]', 'Pogoda_ubiegła_[kod]', 'Chmury_CL_tekstem', 'Chmury_CM_tekstem', 'Chmury_CH_tekstem', 'Chmury_CL_[kod]', 'Chmury_CM_[kod]', 'Chmury_CH_[kod]', 'Stan_gruntu_[kod]', 'Status_pomiaru_HSS', 'Status_pomiaru_GRSN', 'Status_pomiaru_HPRO', 'Status_pomiaru_CIPR']]

df.drop(cols, axis = 1, inplace = True)
```

```python
#uzupełnianinie brakujących danych metodą interpolate
for col in df.columns:
    df[col] = df[col].interpolate(limit_direction='both')
```

```python
#Podział zbioru na zmienne objaśniające i zmienną objaśnianą

y = df['Niepodległości']
X = df.drop('Niepodległości', axis = 1)
```

```python
#Transformacja logarytmiczna
y = np.log(y)
```

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler()
new_X = scaler.fit_transform(X)  #zapisuję jako nową zmienną, bo StandardScaler konwertuje DataFrame do Series
```

```python
#podział na zbiór treningowy i testowy
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_X, y, test_size = 0.3, random_state = 123)
```

```python
from sklearn.model_selection import KFold, cross_val_score
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_squared_log_error, median_absolute_error, r2_score
```

### Podstawowy model

```python
model = xgb.XGBRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
mean_squared_error(y_test, pred)
```

### Optymalizacja

```python
from hyperopt import hp
from hyperopt import hp, fmin, tpe, STATUS_OK, Trials
from functools import partial


def objective_xgb(space):
    
    xgb_params = {
        'max_depth': int(space['max_depth']),
        'colsample_bytree': space['colsample_bytree'],
        'learning_rate': space['learning_rate'],
        'subsample': space['subsample'],
        'seed': int(space['seed']),
        'min_child_weight': int(space['min_child_weight']),
        'reg_alpha': space['reg_alpha'],
        'reg_lambda': space['reg_lambda'],
        'n_estimators': 100
    }
    
    model = xgb.XGBRegressor(**xgb_params)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    score = mean_squared_error(y_test, y_pred)
    
    print("SCORE: {0}".format(score))
    
    return{'loss':score, 'status': STATUS_OK }
    
space ={
    'max_depth': hp.quniform ('x_max_depth', 5, 20, 1),
    'colsample_bytree': hp.uniform ('x_colsample_bytree', 0.8, 1.),
    'learning_rate': hp.uniform ('x_learning_rate', 0.05, 0.2),
    'subsample': hp.uniform ('x_subsample', 0.7, 1.),
    'seed': hp.quniform ('x_seed', 0, 10000, 50),
    'min_child_weight': hp.quniform ('x_min_child_weight', 1, 10, 1),
    'reg_alpha': hp.loguniform ('x_reg_alpha', 0., 1.),
    'reg_lambda': hp.uniform ('x_reg_lambda', 0.7, 1.),
}


trials = Trials()
best_params = fmin(fn=objective_xgb,
            space=space,
            algo=partial(tpe.suggest, n_startup_jobs=1),
            max_evals=20,
            trials=trials)

print("The best params: ", best_params)
```
